Las operaciones esperadas del CRUD son:

Ver agenda
Crear contacto
Consultar contacto
Actualizar contacto
Eliminar contacto

```python


################################################################################
## Diseño y   Construcción:	Elihu Izanami Montoya Estrada
##					izanami49@gmail.com
##					EIME
##	        14 de Febrero del 2026
##          Agenda de Contactos en Python While
################################################################################

#Agenda contactos
agenda = {}

while True:
  print("\nAgenda de Contactos")
  print("1. Agregar Contactos")
  print("2. Buscar Contactos")
  print("3. Listar Contactos")
  print("4. Agregar Contactos")
  print("5. Eliminar Contactos")
  opcion = input("Ingrese una opción: ")

  if opcion == "1":
    nombre = input("Ingrese el nombre del contacto: ")
    telefono = input("Ingrese el número de teléfono: ")
    correo = input("Ingrese el correo electrónico: ").strip()

    agenda[nombre] = {
        "telefono": telefono,
        "correo": correo
    }
    print(f"Contacto {nombre} creado con éxito.")

  elif opcion == "2":
    nombre = input("Ingresa el nombre del contacto que vas a buscar:")
    if nombre in agenda:
      print(f"Nombre: {nombre}")
      print(f"Teléfono: {agenda[nombre]['telefono']}")
      print(f"Correo: {agenda[nombre]['correo']}")
    else:
      print(f"El contacto {nombre} no existe en la agenda.")

  elif opcion == "3":
    print("Lista de Contactos")
    for nombre in agenda:
      print(f"Nombre: {nombre}")
      print(f"Teléfono: {agenda[nombre]['telefono']}")
      print(f"Correo: {agenda[nombre]['correo']}")
      print("*"*20)
    if not agenda:
      print("No hay contactos en la agenda.")

  elif opcion == "4":
    nombre = input("Nombre del contacto a actualizar: ")
    if nombre in agenda:
      telefono = input("Ingrese el nuevo número de teléfono: ")
      correo = input("Ingrese el nuevo correo electrónico: ")
      agenda.update({
          nombre: {
            "telefono": telefono,
            "correo": correo
          }
      })
      print(f"Contacto {nombre} actualizado con éxito.")
    else:
      print(f"El contacto {nombre} no existe en la agenda.")

  elif opcion == "5":
    nombre = input("Ingresa el nombre del contacto que vas a eliminar:")
    if nombre in agenda:
      agenda.pop(nombre)
      print(f"Contacto {nombre} fue eliminado con éxito.")
    else:
      print(f"El contacto {nombre} no existe en la agenda.")

  elif opcion == "6":
    print("Saliendo del programa... BYE .......")
    break
  else:
    print("Opción no válida. Por favor, ingrese una opción válida.")

```
